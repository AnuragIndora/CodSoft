```python
import pandas as pd, numpy as np, matplotlib.pyplot as plt, plotly.express as px, seaborn as sns
```

```python
test_df = pd.read_csv('Dataset/fraudTest.csv').set_index("Unnamed: 0")
test_df.sample(5)
```

```python
col = ["cc_num", "merchant", "first", "last", "street", "city", "state", "zip", "lat", "long", "job", "trans_num", "unix_time", "merch_lat", "merch_long"]
df = test_df.drop(col, axis=1)
df.sample(5)
```

```python
df['trans_date_trans_time'] = pd.to_datetime(df['trans_date_trans_time'])
df['dob'] = pd.to_datetime(df['dob'])

df.info()
```

```python
df['trans_date'] = df['trans_date_trans_time'].dt.date
df['trans_time'] = df['trans_date_trans_time'].dt.time
df['trans_hr'] = df['trans_date_trans_time'].dt.hour
df.info()
```

```python
def time_of_day(data: int) -> object:
    if 0 <= data['trans_hr'] <= 6:
        val = 'night'
    elif 22 <= data['trans_hr'] < 24:
        val = 'night'
    elif 6 <= data['trans_hr']  <= 12:
        val = 'morning'
    elif 12 <= data['trans_hr']  <= 16:
        val = 'afternoon'
    else :
        val = 'evening'
    return val
```

```python
df['time_of_day'] = df.apply(time_of_day, axis=1)
df.sample(5)
```

```python
def age_category(yr: int) -> object:
    if 0< (2024 - yr['birth_year']) <= 15:
        val = 'adult'
    elif 15 < (2024 - yr['birth_year']) <=  29:
        val = 'youth'
    elif 29 < (2024 - yr['birth_year']) <= 59:
        val = 'middle-aged'
    else:
        val = 'old'
    return val
```

```python
df['birth_year'] = df['dob'].dt.year
```

```python
df['Age'] = df.apply(age_category, axis=1)
df.sample(5)
```

```python
df = df.drop(["trans_date_trans_time", "dob", "trans_hr", "birth_year"], axis=1)
df.sample(5)
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
categorical_columns = df.select_dtypes(include=["object"])

le = LabelEncoder()

for col in categorical_columns:
    df[col] = le.fit_transform(df[col])
    # print("Label Encoder Class : ", le.classes_)
```

```python
df.info()
```

```python
# Make Balanced Dataset 

# need to resize it 
from sklearn.utils import resample

# majority 
major_class = df[df['is_fraud'] == 0]
# minority 
minor_class = df[df['is_fraud'] == 1]

# down sample majority class 
majority_downsampled = resample(major_class, replace=False, n_samples=minor_class.shape[0], random_state=36)

df = pd.concat([majority_downsampled, minor_class])

df.shape
```

```python
df.to_csv('Dataset/balance_test.csv', index=True, header=True)
```
